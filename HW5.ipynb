# Plot the Skin Depth of Copper vs Frequency up to 200 GHz

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plot
import math
```

```python
Freq_MIN = 1
Freq_MAX = 200 * (10**9)
NPTS = 100
PI = 22/7
 
rho_copper = 1.678 * (10 ** -8)
permeability_free_space = 4 * PI * (10**-7)
relative_permeability_copper = 0.999991

freq_val = np.linspace(Freq_MIN, Freq_MAX, NPTS)

def skin_depth (freq):
    return math.sqrt (rho_copper / ( PI * freq * permeability_free_space * relative_permeability_copper))

v_skin_depth = np.vectorize(skin_depth)

skin_depth_vals = v_skin_depth(freq_val)

d = {'frequency (Hz)': freq_val, 
     'Skin Depth (m)': skin_depth_vals
    }
data = pd.DataFrame(data=d)
data = data.set_index('frequency (Hz)')

# Set the width and height of the figure
fig = plot.figure(figsize=(14,6))

# Add title
plot.title("Skin Depth vs Frequency Copper")
plot.loglog()

sns.lineplot(data)
```
