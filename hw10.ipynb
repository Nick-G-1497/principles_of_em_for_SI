```python
import numpy as np
import math
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from math import e, pi
import cmath
import matplotlib.pyplot as plt
from numpy import linspace, meshgrid, cos, pi, exp
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

from mpl_toolkits.mplot3d import axes3d

c = 3e8
freq = 1e9
lamda = c/freq
b =  lamda
B = 3
N = 30

mu = 1.26e-6
epsilon = 8.85e-12

omega = 2 * math.pi * freq

def H_z (y, z, n, omega):
    h = n * math.pi /b 
    prop_const = cmath.sqrt (h**2 - omega**2 * mu * epsilon)
    #print(prop_const)
    return B * np.cos(h * y) * math.e ** (-prop_const *z)


Y_VAL = np.linspace(0, b, N)
Z_VAL = np.linspace(0, 3*lamda, N)

# Y1, Z1 = np.meshgrid(Z_VAL, Y_VAL)
Y2, Z2 = np.meshgrid(Y_VAL, Z_VAL)

first_mode = H_z(Y2, Z2, 0, omega)
second_mode = H_z(Y2, Z2, n=1, omega=omega)

# asx, asy, asz = np.ptp(Y_VAL), np.ptp(Z_VAL), np.ptp(first_mode)

ax = plt.figure().add_subplot(projection='3d')
# x = linspace(0,1,64)
# y = linspace(0,20,64)
# X,Y = meshgrid(x,y)
Z = H_z(Y2, Z2, 1, omega)
#X, Y, Z = axes3d.get_test_data(0.05)
#print(Z)


ax.get_proj = lambda: np.dot(Axes3D.get_proj(ax), np.diag([0.5, 2, 0.5, 2]))
# Plot the 3D surface
ax.plot_surface(Y2, Z2, first_mode.real, edgecolor='purple', lw=0.5, rstride=2, cstride=2,
                alpha=0.3)

# Plot projections of the contours for each dimension.  By choosing offsets
# that match the appropriate axes limits, the projected contours will sit on
# the 'walls' of the graph
#ax.contourf(X, Y, Z, zdir='z', offset=-100, cmap='coolwarm')
ax.contourf(Y2, Z2, first_mode.real, zdir='x', offset=-0.25, cmap='rainbow')
#ax.contourf(X, Y, Z, zdir='y', offset=40, cmap='coolwarm')

ax.set(xlim=(0, b), ylim=(0, 3*lamda), zlim=(-B, B),
      xlabel='Y', ylabel='Z', zlabel='Hz')
ax.set_yticklabels([])
ax.set_xticklabels([])
ax.set_zticks([])
ax

ax.grid(False)


plt.show()

```
